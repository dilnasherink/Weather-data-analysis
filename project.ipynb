## Data cleaning

### Clean Numericals columns with characters

```python
df['sunrise']=pd.to_numeric(
    df['sunrise']
    .astype(str)
    .str.replace(r'[^0-9,-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
```

```python
df
```

```python
df["last_updated"] = pd.to_datetime(df["last_updated"],
                                    errors="coerce")
df.info()
```

# Summary statistics

- Mean
- Median
- Mode
- Standard deviation

#### Mean

- The mean is the sum of all values divided by the number of values.


```python
print(df.select_dtypes(include='number').mean())
```

#### Median

- The median is the middle value when data is arranged in order.


```python
print(df.select_dtypes(include='number').median())
```

#### Mode

- The mode is the value that appears most frequently.


```python
c= df.select_dtypes(include='object')
for i in c:
    print(i,df[i].mode())
    print()
```

#### Standard deviation

- It measures how spread out the data is from the mean.

    Low standard deviation → values are close to mean

    High standard deviation → values are spread out


```python
print(df.select_dtypes(include='number').std())
```

- comparing mean,median and standard deviation


```python
stats_df = df.select_dtypes(include='number').agg(['mean','median','std']).T
stats_df
```
