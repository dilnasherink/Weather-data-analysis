### **Histogram**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5)) # to adjust figure size
sns.histplot(data=df_new,x="temp_c",bins=20,color="green")
plt.title("Distribution of temp_c")
plt.xlabel("temperature in celsius")
plt.ylabel("T")
plt.show()

```


* Most temperatures are **between 20°C and 30°C**, showing this is the most common range.
* The distribution is **slightly right-skewed**, meaning there are a few higher temperature values.
* Very low temperatures (below 0°C) and very high ones (above 40°C) are **rare**.



```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = [
    'temp_c', 'wind_kph', 'wind_degree', 'pressure_mb', 'precip_mm',
    'humidity', 'cloud', 'feels_like_celsius', 'visibility_km',
    'uv_index', 'gust_kph', 'o3', 'no2', 'so2',
    'PM_2.5', 'PM_10'
]

for col in num_cols:
    plt.figure(figsize=(5,5))
    sns.histplot(data=df_new, x=col, bins=20, color='teal', kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

insight of the histogram(histplot) graphs :

- **Temperature (temp_c):** Most values cluster between **20–30°C**, with a peak around 30°C. Very cold (<0°C) and very hot (>40°C) readings are rare. The distribution is slightly right‑skewed.  
- **Wind Speed (wind_kph & gust_kph):** Most observations are at lower speeds, but there are a few extreme outliers reaching very high values, showing occasional strong winds.  
- **Wind Degree:** Evenly spread across 0–360°, reflecting all compass directions.  
- **Pressure (pressure_mb):** Strong concentration around **1000 mb**, with a few higher values as anomalies.  
- **Precipitation (precip_mm):** Mostly low rainfall, with occasional spikes indicating heavy rain events.  
- **Humidity & Cloud:** Both show high frequencies near **80–100%**, meaning moist and cloudy conditions are common.  
- **Visibility (km):** Most values are in the **10–20 km** range, suggesting generally clear visibility.  
- **UV Index:** Peaks around **6–10**, but some extreme values up to 14 are present.  
- **Air Quality (O₃, NO₂, SO₂, PM2.5, PM10):** Most pollutants are concentrated at lower levels, though histograms reveal occasional spikes, indicating episodes of poor air quality.  

*Overall Insight:* The histograms show that most weather variables fall within normal ranges, with clear peaks (like temperature around 30°C, pressure near 1000 mb, humidity near 80–100%). Outliers exist for wind, pollutants, and UV index, highlighting rare but significant extreme events.

### **count plot**

- categorical data

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_grouped = df_new.groupby("weather").size().reset_index(name="count")

plt.figure(figsize=(20,5))
sns.barplot(data=df_grouped, x="weather", y="count", color="yellow")

plt.title("Count Plot of Weather Conditions")
plt.xlabel("Conditions of Weather")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

- Most frequent conditions: Weather types like partly cloudy, sunny, and clear dominate the dataset, shown by the tallest bars.

- Moderate frequency: Conditions such as cloudy, mist, and light rain appear often but less than the top three.

- Rare conditions: Extreme events like blizzard, heavy rain, or freezing drizzle occur much less frequently, with very short bars.

- Overall trend: The dataset is heavily skewed toward everyday, stable conditions (sunny/partly cloudy/clear), while severe or unusual weather events are rare.

### **Bar plot**


- categorical vs numerical

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_grouped = df_new.groupby("weather", as_index=False)["humidity"].mean()

plt.figure(figsize=(20,5))
sns.barplot(data=df_grouped, x="weather", y="humidity", color="skyblue")

plt.title("Average Humidity by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Average Humidity")
plt.xticks(rotation=90) 
plt.show()
```



- **High humidity**: Conditions involving precipitation or fog (like heavy rain, drizzle, mist, fog, and overcast) show the highest average humidity, often close to 100%.  
- **Lower humidity**: Clear and sunny conditions have noticeably lower humidity compared to rainy or cloudy weather.  
- **General pattern**: The chart highlights that moisture‑related weather phenomena consistently correspond to higher humidity, while stable, clear skies correspond to lower values.  

-  Humidity levels rise significantly during wet or foggy conditions and drop during clear or sunny weather.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_grouped = df_new.groupby("weather", as_index=False)["o3"].mean()

plt.figure(figsize=(20,6))

sns.barplot(data=df_grouped, x="weather", y="o3", color="green")

plt.title("Average O₃ (ozone) by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Average O₃ Levels")

plt.xticks(rotation=90) 
plt.show()
```



- **Variation across conditions**: Ozone (O₃) levels change noticeably depending on the weather type.  
- **Peaks**: Certain conditions like heavy rain or thunderstorms show higher average O₃ values.  
- **Dips**: Clear, sunny, or calm conditions tend to have lower ozone levels.  
- The graph highlights how atmospheric conditions influence ozone concentration, with unstable or stormy weather linked to higher levels.  

- Ozone levels are not constant—they rise and fall depending on the weather, with stormy conditions generally showing higher concentrations than clear skies.

### **Heat map**

- correlation

```python

corr = df_new[["temp_c", "humidity","wind_degree","wind_kph","uv_index","gust_kph"]].corr()

plt.figure(figsize=(6,6))

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap: Temp vs Feels Like")

plt.show()
```


- **Strongest relationship**: `wind_kph` and `gust_kph` show a very high correlation (0.95). This makes sense because gust speed is essentially a short‑term spike in wind speed.  
- **Moderate positive link**: `temp_c` and `uv_index` correlate at 0.49, suggesting that higher temperatures often coincide with stronger UV radiation.  
- **Temperature vs. humidity**: The correlation is weak (around ‑0.04), meaning humidity doesn’t directly track with temperature in this dataset.  
- **Feels like vs. temp**: As expected, these are highly correlated (close to 1), confirming that perceived temperature is strongly driven by actual temperature.  
- **Other variables**: Wind direction (`wind_degree`) shows little correlation with most factors, indicating it varies independently.

 Overall, the heatmap highlights that **wind speed metrics are tightly linked**, while **temperature and UV share a moderate relationship**, and **humidity behaves independently**.
